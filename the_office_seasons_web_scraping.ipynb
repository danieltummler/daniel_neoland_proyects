```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_The_Office_(American_TV_series)_episodes"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
tables = [table for table in soup.find_all("table", class_ = "wikitable plainrowheaders wikiepisodetable")[:9]]
```

```python
episodes = []
for num, table in enumerate(tables, 1):
    
    rows = [row for row in table.find_all("tr", class_ = "vevent")]
    
    for row in rows:
        
        episodes.append([num, 
                         int(row.find("th").text), 
                         int(row.find("td").text), 
                         row.find("td", class_ = "summary").text.split('"')[1], 
                         row.find_all("td")[2].text, 
                         row.find_all("td")[3].text, 
                         " ".join(row.find_all("td")[4].text.split("\xa0")), 
                         int(row.find_all("td")[5].text), 
                         float((row.find_all("td")[6].text.split("["))[0])])
```

```python
df = pd.DataFrame(episodes, columns = ["season", "no. overall", "no. in season", "title", "directed by",
                                  "written by", "original air date", "prod. code", "u.s. viewers (millions)"])
```
